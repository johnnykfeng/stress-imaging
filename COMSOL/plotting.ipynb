```python
import pandas as pd
import plotly.express as px

# Read the data
df = pd.read_csv('COMSOL_simulation_stresses.csv')

# Create subplots for s1, s2, and theta
s1_range = [df['s1'].quantile(0.05), df['s1'].quantile(0.95)]
fig_s1 = px.scatter(df, x='X', y='Y', color='s1', title='s1-principal stress',
                    color_continuous_scale='jet', range_color=s1_range)
fig_s1.show()

s2_range = [df['s2'].quantile(0.05), df['s2'].quantile(0.95)]
fig_s2 = px.scatter(df, x='X', y='Y', color='s2', title='s2-principal stress',
                    color_continuous_scale='jet', range_color=s2_range)
fig_s2.show()

theta_range = [df['theta'].quantile(0.05), df['theta'].quantile(0.95)]
fig_theta = px.scatter(df, x='X', y='Y', color='theta', title='theta-principal angle',
                      color_continuous_scale='jet', range_color=theta_range)
fig_theta.show()
```
